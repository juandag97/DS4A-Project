```python
Consump['datetime']=pd.to_datetime(Consump['Time Bucket (America/New_York)'])
Consump.drop(columns=['Time Bucket (America/New_York)','Unnamed: 0'],inplace=True)
Consump['Year']=Consump['datetime'].dt.year
Consump['Month']=Consump['datetime'].dt.month
Consump['Day']=Consump['datetime'].dt.day
print(Consump.describe())
```

```python

Consump=Consump.sort_values(by=["Month","Day"]).reset_index(drop=True)
Consump["Year"]=2020
Consump

```

```python
Consump_month=Consump.drop(columns=["datetime","Day"]).groupby(["Month","Year"]).sum()
Consump_month.to_csv('/home/bleon/Documents/DS4A/DS4A_Project/Processed_Data/Consumption_month.csv')
Consump.to_csv('/home/bleon/Documents/DS4A/DS4A_Project/Processed_Data/Consumption_hour.csv')
```
